```python
import pandas as pd
dataframe = pd.read_csv('Total.csv')
dataframe
```

```python
def get_employee_details(employee_name):
    
    try:
    
        result = dataframe[dataframe['EmployeeName'].str.lower() == employee_name.lower()]
        

        if not result.empty:
            return result
        else:
            
            return pd.DataFrame({'Error': [f"No employee found with the name: {employee_name}"]})
    except Exception as e:
        
        return pd.DataFrame({'Error': [f"An error occurred: {str(e)}"]})
    
employee_name = input("Please enter the employee's name: ") # Ask for the employee's name
result = get_employee_details(employee_name)
display(result)
```

```python

try:
    dataframe = pd.read_csv('Total.csv')
except FileNotFoundError:
    print("Error: 'Total.csv' not found.")
    dataframe = pd.DataFrame()


required_columns = ['EmployeeName', 'BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay']
if not dataframe.empty and set(required_columns).issubset(dataframe.columns):
    
    dataframe['Benefits'] = pd.to_numeric(dataframe['Benefits'], errors='coerce').fillna(0)
    dataframe[['BasePay', 'OvertimePay', 'OtherPay', 'TotalPay']] = dataframe[['BasePay', 'OvertimePay', 'OtherPay', 'TotalPay']].apply(pd.to_numeric, errors='coerce').fillna(0)

    
    def get_employee_salary(employee_name):
        employee_data = dataframe[dataframe['EmployeeName'] == employee_name]
        if not employee_data.empty:
            return employee_data[['EmployeeName', 'BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay']]
        else:
            return pd.DataFrame({'Error': [f"No employee found with the name: {employee_name}"]})


    def get_all_employee_salaries():
        return dataframe[['EmployeeName', 'BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay']]

    
    result = get_employee_salary('NATHANIEL FORD')
    display(result)
```

```python
import zipfile


def export_to_csv_and_zip():
    try:
        if dataframe.empty:
            raise ValueError("No data to export.")
        
        all_data = get_all_employee_salaries()
        
        zip_name = "Employee Profile.zip"
        csv_name = "employee_details.csv"
        
        with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
            
            zipf.writestr(f"Employee Profile/{csv_name}", all_data.to_csv(index=False))
        print(f"Employee details exported and zipped to {zip_name} within 'Employee Profile'")
    except Exception as e:
        print(f"Error during export and zipping: {str(e)}")

export_to_csv_and_zip()
```
